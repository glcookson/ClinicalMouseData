```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combind_df = pd.merge(clinical_trial_df, mouse_drug_df, on="Mouse ID", how="left")

# Display the data table for preview
combind_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_drug_time = combind_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean()
# Convert to DataFrame
mean_tumor_drug_time_df = pd.DataFrame(mean_tumor_drug_time)
# Preview DataFrame
mean_tumor_drug_time_df.reset_index()
mean_mung_df = mean_tumor_drug_time_df.reset_index()
mean_mung_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_drug_time = combind_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
sem_tumor_drug_time_df = pd.DataFrame(sem_tumor_drug_time)
# Preview DataFrame
sem_tumor_drug_time_df= sem_tumor_drug_time_df.reset_index()
sem_tumor_drug_time_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
refited_df = mean_mung_df.pivot(index="Timepoint", columns="Drug")['Tumor Volume (mm3)']
refited_df

# Preview that Reformatting worked
refited_df
```

```python
# Generate the Plot (with Error Bars)

format_drug_ls = [('Capomulin', 'o', 'red'), ('Infubinol', '^', 'blue'), ('Ketapril', 's', 'green'), ('Placebo', 'd', 'black')]
for d,m,c in format_drug_ls:
    df = mean_mung_df.loc[mean_mung_df["Drug"] == d]
    de = sem_tumor_drug_time_df.loc[sem_tumor_drug_time_df["Drug"] == d]
    plot = plt.errorbar(df["Timepoint"], df["Tumor Volume (mm3)"], yerr=de["Tumor Volume (mm3)"], label=d, marker=m, color=c)
plot

plt.legend(loc="best")
plt.title("Tumor Response to Treament")
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.ylim(ymin = 35, ymax = 70)
plt.grid()


#Save the Figure
# Show the Figur
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_Metastatic_Sites = combind_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean()
# Convert to DataFrame
mean_Metastatic_Sites_df = pd.DataFrame(mean_Metastatic_Sites)
# Preview DataFrame
mean_Metastatic_Sites_df = mean_Metastatic_Sites_df.reset_index()
mean_Metastatic_Sites_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_Metastatic_Sites = combind_df.groupby(['Drug', 'Timepoint'])[['Metastatic Sites']].sem()
# Convert to DataFrame
sem_Metastatic_Sites_df = pd.DataFrame(sem_Metastatic_Sites)
# Preview DataFrame
sem_Metastatic_Sites_df = sem_Metastatic_Sites_df.reset_index()
sem_Metastatic_Sites_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
refited_2df = mean_Metastatic_Sites_df.pivot(index="Timepoint", columns="Drug")['Metastatic Sites']
refited_2df
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)
format_drug_ls = [('Capomulin', 'o', 'red'), ('Infubinol', '^', 'blue'), ('Ketapril', 's', 'green'), ('Placebo', 'd', 'black')]
for d,m,c in format_drug_ls:
    df = mean_Metastatic_Sites_df.loc[mean_Metastatic_Sites_df["Drug"] == d]
    de = sem_Metastatic_Sites_df.loc[sem_Metastatic_Sites_df["Drug"] == d]
    plot = plt.errorbar(df["Timepoint"], df["Metastatic Sites"], yerr=de["Metastatic Sites"], label=d, marker=m, color=c)
plot

plt.legend(loc="best")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.ylim(ymin = 0, ymax = 3.5)
plt.grid()


# Save the Figure

# Show the Figure
plt.show()
```
